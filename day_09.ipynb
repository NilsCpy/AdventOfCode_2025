```python
data = open('data/day_09.txt', 'r').read().split('\n')[:-1]
```

```python
data_clean = [list(map(int, elem.split(','))) for elem in data]
```

## Part 1

```python
def get_area(ptA, ptB):
    return (abs(ptA[0] - ptB[0])+1) * (abs(ptA[1] - ptB[1])+1)
```

```python
area = []
for i, ptA in enumerate(data_clean):
    for j, ptB in enumerate(data_clean[:i]):
        area.append((i, j, get_area(ptA, ptB)))

areas_sorted = sorted(area, key = lambda x:x[2])
```
