```python
data_clean = [list(map(int, elem.split(','))) for elem in data]
```

```python
def get_distance(ptA, ptB):
    return ((ptB[0] - ptA[0])**2 + (ptB[1] - ptA[1])**2 + (ptB[2] - ptA[2])**2)**0.5
```

```python
# (id_ptA, id_ptB, distance)

distances = []
for i, ptA in enumerate(data_clean):
    for j, ptB in enumerate(data_clean[:i]):
        distances.append((i, j, get_distance(ptA, ptB)))

distances_sorted = sorted(distances, key = lambda x:x[2])
```

## Part 1

```python
points = {}
nb_groups = 0

for junc in range(1000):
    dist = distances_sorted[junc]

    if dist[0] in points and not(dist[1] in points):
        points[dist[1]] = points[dist[0]]

    elif dist[1] in points and not(dist[0] in points):
        points[dist[0]] = points[dist[1]]

    elif not (dist[0] in points) and not(dist[1] in points):
        points[dist[0]] = nb_groups
        points[dist[1]] = nb_groups
        nb_groups += 1

    elif dist[0] in points and dist[1] in points:
        old = points[dist[1]]
        for key in points.keys():
            if points[key] == old:
                points[key] = points[dist[0]]
```

```python
groups = {}

for key in points.keys():
    if points[key] in groups.keys():
        groups[points[key]].append(key)
    else:
        groups[points[key]] = [key]
```

```python
groups_sorted = dict(sorted(groups.items(), key=lambda x:len(x[1])))
```

```python
total = 1

for group in list(groups_sorted.keys())[-3:]:
    total *= len(groups_sorted[group])

total
```
