```python
data = open('data/day_04.txt', 'r').read().split('\n')[:-1]
```

## Part 1

```python
max_x = len(data[0])-1
max_y = len(data)-1

count = 0

for x in range(max_x+1):
    for y in range(max_y+1):
        if data[y][x] != '@':
            continue
        if [data[j][i] for j in range(max(0, y-1), min(y+1, max_y)+1) for i in range(max(0, x-1), min(x+1, max_x)+1)].count('@') < 5:
            count += 1

count
```

## Part 2

```python
data_proc = []

max_x = len(data[0])-1
max_y = len(data)-1

count = 0

while data_proc != data:

    data_proc = data.copy()

    pos = []

    for x in range(max_x+1):
        for y in range(max_y+1):

            if data[y][x] != '@':
                continue

            if [data[j][i] for j in range(max(0, y-1), min(y+1, max_y)+1) for i in range(max(0, x-1), min(x+1, max_x)+1)].count('@') < 5:
                pos.append((x, y))
                count += 1

    for elem in pos:
        line = list(data_proc[elem[1]])
        line[elem[0]] = '.'
        data_proc[elem[1]] = ''.join(line)

count

## ça ne marche plus mais ça a marché je jure
```
