```python
data = open('data/day_06.txt', 'r').read().split("\n")[:-1]
```

## Part 1

```python
data_clean = [elem.replace("      ", ",").replace("     ", ",").replace("    ", ",").replace("   ", ",").replace("  ", ",").replace(" ", ",").strip(',') for elem in data]
```

```python
data_clean = [elem.split(',') for elem in data_clean]
```

```python
total = 0

for ind, symb in enumerate(data_clean[-1]):
    if symb == '+':
        total += int(data_clean[0][ind]) + int(data_clean[1][ind]) + int(data_clean[2][ind]) + int(data_clean[3][ind])
    elif symb == '*':
        total += int(data_clean[0][ind]) * int(data_clean[1][ind]) * int(data_clean[2][ind]) * int(data_clean[3][ind])

total
```

## Part 2

```python
new_data = [elem[::-1] for elem in data]
```

```python
total = 0

ind_sign = 0
list_sign = list(new_data[-1].replace(' ', ''))

i = 0
while i < len(new_data[0]):

    subtotal = 0
    if list_sign[ind_sign] == '*':
        subtotal = 1

    a, b, c, d = new_data[0][i], new_data[1][i], new_data[2][i], new_data[3][i]

    while ((a, b, c, d) != (' ', ' ', ' ', ' ')) and (i < len(new_data[0])):
        if list_sign[ind_sign] == '+':
            subtotal += int(a+b+c+d)
        elif list_sign[ind_sign] == '*':
            subtotal *= int(a+b+c+d)
        i += 1
        if i != 3752:
            a, b, c, d = new_data[0][i], new_data[1][i], new_data[2][i], new_data[3][i]

    ind_sign += 1
    total += subtotal
    i+=1

total
```
