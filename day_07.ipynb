```python
data = open('data/day_07.txt', 'r').read().split("\n")[:-1]
data_used = data.copy()
```

## Part 1

```python
data_used = [list(line) for line in data]
```

```python
activated = []

for indline, line in enumerate(data_used):
    if indline == 0:
        continue
    for indelem, elem in enumerate(line):
        if elem == '.' and data_used[indline-1][indelem] in ['S', '|']:
            data_used[indline][indelem] = '|'
        if elem == '^' and data_used[indline-1][indelem] == '|':
            activated.append((indline, indelem))
            if indelem > 0:
                data_used[indline][indelem-1] = '|'
            if indelem < len(line)-1:
                data_used[indline][indelem+1] = '|'

len(activated)
```

## Part 2

```python
data_used = [list(line) for line in data]
```

```python
for indline, line in enumerate(data_used):
    if indline == 0:
        continue
    for indelem, elem in enumerate(line):

        if isinstance(elem, int) and isinstance(data_used[indline-1][indelem], int):
            data_used[indline][indelem] = elem + data_used[indline-1][indelem]

        if elem == '.' and data_used[indline-1][indelem] == 'S':
            data_used[indline][indelem] = 1
        if elem == '.' and isinstance(data_used[indline-1][indelem], int):
            data_used[indline][indelem] = data_used[indline-1][indelem]

        if isinstance(elem, int) and isinstance(data_used[indline-1][indelem], int):
            data_used[indline][indelem] = elem + data_used[indline-1][indelem]

        if elem == '^' and isinstance(data_used[indline-1][indelem], int):
            if indelem > 0:
                if isinstance(data_used[indline][indelem-1], int):
                    data_used[indline][indelem-1] = data_used[indline-1][indelem] + data_used[indline][indelem-1]
                if data_used[indline][indelem-1] == '.':
                    data_used[indline][indelem-1] = data_used[indline-1][indelem]

            if indelem < len(line)-1:
                if isinstance(data_used[indline][indelem+1], int):
                    data_used[indline][indelem+1] = data_used[indline-1][indelem] + data_used[indline][indelem+1]
                if data_used[indline][indelem+1] == '.':
                    data_used[indline][indelem+1] = data_used[indline-1][indelem]
```

```python
total = 0

for elem in data_used[-1]:
    if isinstance(elem, int):
        total += elem

total
```
