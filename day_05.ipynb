```python
data = open('data/day_05.txt', 'r').read().split("\n\n")

ranges = data[0].split('\n')
ids = data[1].split('\n')[:-1]
```

## Part 1

```python
count = 0

for id in ids:

    intid = int(id)

    for ran in ranges:

        l, u = map(int, ran.split('-'))

        if l <= intid and intid <= u:

            count += 1
            break

count
```

## Part 2

```python
good_ranges = []
for ran in ranges:
    l, u = map(int, ran.split('-'))
    good_ranges.append((l, u))
```

```python
def main2(ranges):

    sortranges = sorted(ranges, key = lambda x:x[0])

    result = [sortranges[0]]

    for elem in sortranges:
        if elem[1] <= result[-1][1]:
            continue
        elif elem[0] <= result[-1][1]:
            result.append((result[-1][1]+1, elem[1]))
        else:
            result.append(elem)

    total = 0

    for elem in result:
        total += (elem[1]-elem[0]+1)

    return total
```

```python
main2(good_ranges)
```
