```python
data = open('data/day_03.txt', 'r').read().split('\n')[:-1]
```

## Part 1

```python
total = 0

for line in data:
    first = max(line[:-1])
    index = line.index(first)
    second = max(line[index+1:])
    total += int(first+second)

total
```

## Part 2

```python
total = 0

for line in data:

    numb = ''
    index = -1

    line += '0'

    for i in range(12):

        digit = max(line[index+1:-(12-i)])
        index += line[index+1:-(12-i)].index(digit) + 1
        numb += digit

    total += int(numb)

total
```
