```python
data = open('data/day_01.txt', 'r').read()
```

## Part 1

```python
# flex

[sum([int(elem) for elem in data.replace("L", "-").replace("R", "").split('\n')[:-1]][:ind+1])%100 for ind in range(len(data.split("\n"))-1)].count(50)
```

```python
# better

data_clean = [int(elem) for elem in data.replace("L", "-").replace("R", "").split('\n')[:-1]]

total = 50
count = 0

for elem in data_clean:
    total += elem
    total = total % 100
    if total == 0:
        count += 1

print(count)
```

## Part 2

```python
# Solution de fragile mais j'en ai marre

data_clean = [int(elem) for elem in data.replace("L", "-").replace("R", "").split('\n')[:-1]]

total = 50
count = 0

for elem in data_clean:
    sign = 1
    if elem < 0:
        sign = -1
    for i in range(abs(elem)):
        total += sign
        if total % 100 == 0:
            count += 1

count
```
