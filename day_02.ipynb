```python
data = open('data/day_02.txt', 'r').read()[:-1].split(',')
```

## Part 1

```python
def is_double(numb):

    numstr = str(numb)
    length = int(len(numstr))

    if length%2 == 1:
        return False

    if numstr[:length//2] == numstr[length//2:]:
        return True

    return False
```

```python
total = 0

for rang in data:
    start, end = rang.split('-')
    start, end = int(start), int(end)
    for number in range(start, end+1):
        if is_double(number):
            total += number

total
```

## Part 2

```python
def is_multiple(numb):

    numstr = str(numb)
    length = int(len(numstr))

    sizes = [size for size in range(1, length) if (length/size) == int(length/size)]

    for size in sizes:

        substrs = [numstr[size*i:size*(i+1)] for i in range(length//size)]
        if len(set(substrs)) == 1:
            return True

    return False
```

```python
total = 0

for rang in data:
    start, end = rang.split('-')
    start, end = int(start), int(end)
    for number in range(start, end+1):
        if is_multiple(number):
            total += number

total
```
